```python
import mc_numpy as mc 
import matplotlib.pyplot as plt

# for argon as a liquid
reduced_temperature = 0.869
density = 26.151
sigma = 3.4e-10
reduced_density = density * 6.022e23 * sigma**3 * 1000
steps, energies, bins, rdf = mc.run_sim(reduced_temperature, reduced_density)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(steps, energies)
plt.xlabel('steps')
plt.ylabel('energies')
plt.ylim([-1000, 1000])
plt.xlim([0, 20000])
plt.title('Energies for Argon as a Liquid')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(bins, rdf)
plt.xlabel('bins')
plt.ylabel('radial distribution function')
plt.title('RDF for Argon as a Liquid')
```
